**Problem Statement:**
The term unemployment is measured using unemployment rate which is the number of people who don't work or earn as a percentage of the total labour force. Due to Covid-19, we have seen tremendous amount of unemployment. That's why according to me this can be a good data science project.

1. Let's start with importing required Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

2. We will be loading Unemployment in India's dataset

```python
data = pd.read_csv('/content/Unemployment_in_India.csv')
data
```

3. We will be checking the information of dataset

```python
data.info()
```

4. We will be describing the dataset

```python
data.describe()
```

5.Now let's check whether the dataset consists any null/missing values

```python
data.isnull().sum()
```

6. As we can see there are several null values for different attributes, we need to drop (remove) the null values.

```python
data = data.dropna()
```

7. We are renaming the columns

```python
data.columns = ['States','Date','Frequency','Estimated Unemployment Rate','Estimated Employed','Estimated Labour Participation Rate', 'Area']
```

8. To view top 5 rows of our dataset

```python
data.head()
```

Checking **Correlation** between the feature of Dataset

```python
data = data.select_dtypes(include=["number"])
```

```python
# Plotting a correlation heatmap
# Set tick parameters using tick_params()
# Computer the correlation matrix and plot the heatmap

plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(8,6))
ax = sns.heatmap(data.corr(), annot=True, cmap='coolwarm', linewidth=3 )

ax.tick_params(size=10, color='w', labelsize=10, labelcolor='w')
ax.set_title('Correlation Heatmap between the features of dataset')
plt.show()
```

**Estimated number of employee** according to different areas of India

```python
# Plotting histplot

data = pd.read_csv("/content/Unemployment_in_India.csv")
data.columns = ['States', 'Date','Frequency','Estimated Unemployment Rate','Estimated Employed','Estimated Labour Participation Rate', 'Area']
plt.title('Indian Unemployment')
sns.histplot(x='Estimated Employed', hue='Area', data=data)
plt.show()
```

**Unemployment rate** according to different areas of India

```python
# Plotting histplot

data.columns = ['States', 'Date','Frequency','Estimated Unemployment Rate','Estimated Employed','Estimated Labour Participation Rate', 'Area']
plt.title('Indian Unemployment')
sns.histplot(x='Estimated Unemployment Rate', hue='Area', data=data)
plt.show()
```

**DASHBOARD**
to analyze the unemployment rate of each Indian state

```python
# Plotting boxplot
data = pd.read_csv("/content/Unemployment_in_India.csv")
data.columns = ['States', 'Date','Frequency','Estimated Unemployment Rate','Estimated Employed','Estimated Labour Participation Rate', 'Area']
plt.figure(figsize=[40,20])
plt.title('Indian Unemployment')
sns.boxplot(x='States', y='Estimated Employed', hue='States', data=data)
plt.show()
```
